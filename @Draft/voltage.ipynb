# Voltage vs Time analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the voltage.csv dataset
df = pd.read_csv('voltage.csv')
df.head()

# convert the Time stamp column to datetime and the State column to float
df['Time stamp'] = pd.to_datetime(df['Time stamp'])

# convert the State column to float, drop non-numeric rows
df['State'] = pd.to_numeric(df['State'], errors='coerce')
df = df.dropna(subset=['State'])

df.head()
```

```python
# plot the voltage over time
plt.figure(figsize=(14, 6))
# use a very thin line to make the plot look nicer
plt.plot(df['Time stamp'], df['State'], linewidth=0.2)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.grid()
plt.title('Voltage over time - ALTA CRDOBA')
plt.show()
```

```python
# compute the mean and the median voltage
vmean = df['State'].mean()
vmedian = df['State'].median()

# Make a histogram of the voltage, with 50 bins
plt.figure(figsize=(14, 6))
plt.hist(df['State'], bins=50)

# add the mean and median voltage as a vertical line
plt.axvline(vmean, color='red', linestyle='dashed', linewidth=1)
plt.axvline(vmedian, color='green', linestyle='dashed', linewidth=1)

plt.xlabel('Voltage')
plt.ylabel('Frequency')
plt.title('Voltage histogram - ALTA CRDOBA')

# draw an horizontal grid
plt.grid(axis='y')

plt.legend({'Promedio':vmean, 'Mediana':vmedian})

plt.show()
```
